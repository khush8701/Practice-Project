```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_rows',None)  #setting the display option to max
pd.set_option('display.max_columns',None)
amz=pd.read_csv(r'F:\Internship - Data Science\DS - Assignment Part 2 data set\amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
fpk=pd.read_csv(r'F:\Internship - Data Science\DS - Assignment Part 2 data set\flipkart_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
result=pd.merge(amz,
               fpk[['uniq_id','retail_price','discounted_price']],
               on='uniq_id')
result.head()
```
